```python
import numpy as np
from matplotlib import pyplot as plt
import math
from matplotlib.lines import Line2D
```

## External influence model

```python
def external_f(a, t):
    innovators = 1000 * (1 - (np.exp(-a*t)))
    return innovators
```

```python
t = np.linspace (0, 50000, 1000)
```

```python
fig, ax = plt.subplots()
ax.plot(t, external_f(0.01, t), color='r')
ax.plot(t, external_f(0.001, t), color='g')
ax.plot(t, external_f(0.0001, t), color='b')

custom_lines = [Line2D([0], [0], color= 'r', lw=4),
                Line2D([0], [0], color='g', lw=4),
                Line2D([0], [0], color='b', lw=4)]

plt.title('External Influence Model')
plt.ylabel('population of innovators')
plt.xlabel('iterations')


ax.legend(custom_lines, ['a=0.01', 'a=0.001', 'a=0.0001'])
plt.show()
```

## Internal Influence Model

```python
def internal_f(b, t):
    innovators = 1000 / (1 + ((950/50) * (np.exp(-b*1000*t))))
    return innovators
```

```python
t = np.linspace (0, 800, 800)
```

```python
fig, ax = plt.subplots()
ax.plot(t, internal_f(0.001, t), color='r')
ax.plot(t, internal_f(0.0001, t), color='g')
ax.plot(t, internal_f(0.00001, t), color='b')

custom_lines = [Line2D([0], [0], color= 'r', lw=4),
                Line2D([0], [0], color='g', lw=4),
                Line2D([0], [0], color='b', lw=4)]

plt.title('Internal Influence Model')
plt.ylabel('population of innovators')
plt.xlabel('iterations')


ax.legend(custom_lines, ['B=0.001', 'B=0.0001', 'B=0.00001'])
plt.show()
```

## Mixed Influence Model

```python
def mixed_f(a, b, t):
    numerator = 1000 - (((a*990) / (a + b*10)) * np.exp(-(a + (b*1000)) * t))
    denominator = 1 + (((b*990) / (a + b*10)) * np.exp(-(a + (b*1000)) * t))
    innovators = numerator / denominator
    return innovators
```

```python
t = np.linspace (0, 800, 800)
```

```python
fig, ax = plt.subplots()
ax.plot(t, mixed_f(0.01, 0.001, t), color='C0')
ax.plot(t, mixed_f(0.01, 0.0001, t), color='C1')
ax.plot(t, mixed_f(0.01, 0.00001, t), color='C2')

ax.plot(t, mixed_f(0.001, 0.001, t), color='C3')
ax.plot(t, mixed_f(0.001, 0.0001, t), color='C4')
ax.plot(t, mixed_f(0.001, 0.00001, t), color='C5')

ax.plot(t, mixed_f(0.0001, 0.001, t), color='C6')
ax.plot(t, mixed_f(0.0001, 0.0001, t), color='C7')
ax.plot(t, mixed_f(0.0001, 0.00001, t), color='C8')

custom_lines = [Line2D([0], [0], color='C0', lw=4), Line2D([0], [0], color='C1', lw=4), Line2D([0], [0], color='C2', lw=4),
                Line2D([0], [0], color='C3', lw=4), Line2D([0], [0], color='C4', lw=4), Line2D([0], [0], color='C5', lw=4),
                Line2D([0], [0], color='C6', lw=4), Line2D([0], [0], color='C7', lw=4), Line2D([0], [0], color='C8', lw=4)]

plt.title('Mixed Influence Model')
plt.ylabel('population of innovators')
plt.xlabel('iterations')


ax.legend(custom_lines, ['a=0.01, B=0.001', 'a=0.01, B=0.0001', 'a=0.01, B=0.00001', 'a=0.001, B=0.001', 'a=0.001, B=0.0001', 'a=0.001, B=0.00001',
                        'a=0.0001, B=0.001', 'a=0.0001, B=0.0001', 'a=0.0001, B=0.00001'])
plt.show()
```
