```python
def google_matrix(G, alpha=0.85):
    M = np.asmatrix(nx.to_numpy_array(G))
    N = len(G)
    if N == 0:
        return M

    # Personalization vector
    p = np.array([0.3,0.2,0.1,0.1,0.1,0.4])
    print(p)
    
    
    # Dangling nodes
    dangling_weights = p
    dangling_nodes = np.where(M.sum(axis=1) == 0)[0]

    # Assign dangling_weights to any dangling nodes 
    # (nodes with no out links)
    for node in dangling_nodes:
        M[node] = dangling_weights

    M /= M.sum(axis=1)  # Normalize rows to sum to 1

    return alpha * M + (1 - alpha) * p
```

```python
def pagerank_numpy(G, alpha=0.85):
    if len(G) == 0:
        return {}
    M = google_matrix(G, alpha)
    
    eigenvalues, eigenvectors = np.linalg.eig(M.T)
    ind = np.argmax(eigenvalues)

    # eigenvector of largest eigenvalue is at ind, normalized
    largest = np.array(eigenvectors[:, ind]).flatten().real
    norm = float(largest.sum())
    return dict(zip(G, map(float, largest / norm)))
```

```python
G3 = nx.Graph()

G3.add_nodes_from(['v1','v2','v3','v4','v5','v6'])

G3.add_edge('v1', 'v2', weight = 7)
G3.add_edge('v1', 'v3', weight = 8)
G3.add_edge('v2', 'v5', weight = 4)
G3.add_edge('v2', 'v3', weight = 3)
G3.add_edge('v3', 'v4', weight = 1)
G3.add_edge('v2', 'v4', weight = 2)
G3.add_edge('v4', 'v6', weight = 1)
G3.add_edge('v5', 'v6', weight = 1)
```

```python
pos = nx.spring_layout(G3, seed=0)

edge_labels = dict([((u,v,),d['weight']) for u,v,d in G3.edges(data=True)])
nx.draw_networkx_edge_labels(G3, pos, edge_labels = edge_labels)
nx.draw_networkx_labels(G3, pos)

nx.draw(G3,pos, connectionstyle='angle3')
```

```python
nodes = list(G3.nodes())

paths = dict(nx.all_pairs_dijkstra_path(G3, 6))
print(paths)
```

```python
for vertex in paths.keys():
    # All initial keys; i.e., v1, v2... vn
    print('Vertex: ', vertex)

    # Iterate through all initial keys... aka, iterate through second dictionary
    for source, connections in paths[vertex].items():
        print('Source: ', source, ' Connections: ', connections)

    print()

```
